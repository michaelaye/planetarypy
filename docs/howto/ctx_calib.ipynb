# How-To: Calibrated CTX data

This is a brief how-to showing how the automatically handled raw data can be calibrated using ISIS and how the `planetarypy` module can help working with the calibrated data.

```python
# import warnings
import warnings

import rasterio

from planetarypy.instruments.mro.ctx import Calib, Raw, CTXCollection

try:
    pass
except KeyError:
    warnings.warn("kalasiris has a problem initializing ISIS")
```

```python
warnings.filterwarnings("ignore", category=rasterio.errors.NotGeoreferencedWarning)
```

```python
pid = "B03_010694_1868_XI_06N285W"
pid = "D20_034983_1833_XN_03N284W"
```

```python
raw = Raw(pid)
raw
```

```python
raw.path
```

```python
cal = Calib(pid, destripe_to_calib=True)
```

```python
cal.cub_path.exists()
```

```python
coll = CTXCollection(list_of_pids=[pid])
coll.calibs
coll.cub_paths
coll.cal_paths
coll.map_paths
coll.geodataframe_path
```

```python
coll.write_isis_filelist(level=1)
```

```python
cal.pipeline()
```

```python
cal.cal_name
```

```python
cal.spatial_summing
```

```python
cal.isis_import()
```

```python
cal.spice_init()
```

```python
cal.calibrate(refresh=True)
```

```python
cal.destripe()
```

```python
cal.cal_path
```

```python
cal.plot_any(cal.cal_path)
```
